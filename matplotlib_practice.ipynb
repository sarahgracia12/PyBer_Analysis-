```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create a plot line with ax.plot()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#Create the plot 
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

#Create labels for the x and y  axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0,45)

# Create a title 
plt.title("PyBer Fare by Month")

# Add a grid 
plt.grid()

#Add the legend 
plt.legend()

```

```python
# Create a chart using the objected-oriented approach 

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker = "*", color = "green", linewidth = 4,)

# Add a legend 
ax.legend("Boston")

# add title and axes labels 
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

#Add y-axis limit 
ax.set_ylim(0,45)

# Add grid lines 
ax.grid()
```

```python
# Create a vertical bar chart 
plt.bar(x_axis, y_axis)
```

```python
# Create the bar chart plot 
plt.bar(x_axis, y_axis, color = "green", label = 'Boston')

# Create labels for the x and y axes 
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title 
plt.title("Pyber Fare by Month")

#Add the legend
plt.legend()
```

```python
# Create a horizontal bar chart 
plt.barh(x_axis, y_axis, color = "magenta", label = 'Boston')
plt.gca().invert_yaxis()

# Add title and label axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")

# Add legend 
plt.legend()
```

```python
# Create a bar chart with object orientation 
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create a horizontal bar chart 
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label = 'Chicago')

ax.set_title("Pyber Fare by Month")
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.legend()

# invert y-axis 
plt.gca().invert_yaxis()
```

```python
# Create a scatter plot 
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create a scatter plot with second method 
plt.scatter(y_axis, x_axis, color = 'red', label = "Chicago")

# Add title and label axes
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Add limit 
plt.xlim(0,50)

#invert x-axis
plt.gca().invert_yaxis()
```

```python
# Create a bubble chart 
plt.scatter(x_axis, y_axis, s = y_axis_larger)

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
# Using list comprehension

plt.scatter(x_axis, y_axis, s= [i * 3 for i in y_axis])
```

```python
# Create a scatter plot using object orientation 
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a bubble chart using object orientation 
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], color = "skyblue", alpha = 0.2, edgecolors = 'black', linewidth = 2, label = 'Boston')

ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

ax.legend()
ax.set_xlim(0,50)
plt.gca().invert_yaxis()
```

```python
# Create a pie chart 
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
# Exploding the pie chart 
plt.subplots(figsize = (8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```
